```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
heart=pd.read_csv('E:/heart.csv')
heart.head()
```

## Analysing Data

```python
heart.isnull().sum()
```

```python
# No null value is present
```

```python
cp1=pd.get_dummies(heart['cp'],drop_first=True)
cp1.head()
```

```python
slope1=pd.get_dummies(heart['slope'],drop_first=True)
slope1.head()
```

```python
thal1=pd.get_dummies(heart['thal'],drop_first=True)
thal1.head()
```

```python
heart.drop(["cp","slope","thal"],axis=1,inplace=True)
heart.head()
```

```python
heart=pd.concat([heart,cp1,slope1,thal1],axis=1)
heart.head()
```

## Using Naves Bayes Classifier

```python
from sklearn.model_selection import train_test_split

y=heart["target"]
X=heart.drop(['target','trestbps'],axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=20)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
algo=GaussianNB()
```

```python
algo.fit(X_train,y_train)
```

```python
y_predict=algo.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_predict)
```
